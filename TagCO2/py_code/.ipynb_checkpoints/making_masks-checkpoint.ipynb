# Create xarray region mask

In this tutorial we will show how to create a mask for arbitrary latitude and longitude grids using xarray. It is very similar to the tutorial Create Mask (numpy).

Import regionmask and check the version:

```python
import regionmask
regionmask.__version__
```

Load xarray and the tutorial data:

```python
import xarray as xr
import numpy as np
from functions import area_latlon
```

```python
fname = '../nc/ODIAC_CO2.monthly.generic.1x1.nc'
ds = xr.open_dataset(fname)
ds
```

```python
# take last year, 2014
dr = ds['CO2'][168:180,:,:]
# calculate time in seconds for each month, and convert unit to kg/m2
for count, time in enumerate(dr['time'].values):
    tmp = np.datetime64(ds['time'][count+1].values) - np.datetime64(ds['time'][count].values)
    time_dif = tmp/np.timedelta64(1, 's')
    dr.values[count,:,:] = dr.values[count,:,:] * time_dif *1e3 # g/m2

#dr_mean = dr.mean(dim='time') # 2014 annual mean, g/m2
dr
```

```python
# calculate grid area (using the above defined function) and convert unit to kg/year
for count, lat in enumerate(ds['lat'].values):
    area = area_latlon(lat1=lat, lat2=lat+1, lon1=10, lon2=11) # km^2
    dr.values[:,count,:] = dr.values[:,count,:]*area*1e-3 # Gg/grid cell, monthly

dr_total = dr.sum(dim='time') # 2014 annual total, Gg/grid cell
dr_total
```

```python
from gamap_colormap import WhGrYlRd  # import the "WhGrYlRd" variable from my custom Python script
type(WhGrYlRd)  # our custom colormap
```

If the longitude and latitude in the xarray object are not called `lon` and `lat`, respectively; their name can be given via the `lon_name` and `lat_name` keyword. Here we use the Giorgi regions.

```python
#mask = regionmask.defined_regions.natural_earth.countries_110.mask(dr_total)
mask = regionmask.defined_regions.giorgi.mask(dr_total)
```

We want to select the country 'GB'. Thus we first need to find out which number this is:

```python
a = mask.max().values
a
```

```python
fname = '../../../../../Desktop/SE_Asia_mask.generic.1x1.nc'
ds = xr.open_dataset(fname)
ds
#np.datetime64('2000-01-01')
```

```python
for count in np.arange(a)+1:
    target = xr.Dataset({"MASK": (("time", "lat", "lon"), np.zeros(180*360).reshape(1,180,360))},coords=ds.coords)
    target = target.astype(dtype='float32')
    mask_TF = mask == count
    target['MASK'][0,:,:] = mask_TF.where(True)
    target['MASK'].attrs = ds['MASK'].attrs
    target['lon'].attrs = ds['lon'].attrs
    target['lat'].attrs = ds['lat'].attrs
    #print(target)
    target.attrs = ds.attrs
    target.attrs['comment'] = 'giorgi region masks, by m.sadiq 2020'
    target.to_netcdf('../nc/MASK' + str(count.astype(int)) + '_giorgi_1x1.nc')
```

```python
target = xr.Dataset({"MASK": (("time", "lat", "lon"), np.zeros(180*360).reshape(1,180,360))},coords=ds.coords)
target = target.astype(dtype='float32')
mask_nan = np.isnan(mask)
target['MASK'][0,:,:] = mask_nan.where(True)
target['MASK'].attrs = ds['MASK'].attrs
target['lon'].attrs = ds['lon'].attrs
target['lat'].attrs = ds['lat'].attrs
#print(target)
target.attrs = ds.attrs
target.attrs['comment'] = 'giorgi region masks, by m.sadiq 2020'
target.to_netcdf('../nc/MASK22_giorgi_1x1.nc')
```

## Select using `where`

`xarray` provides the handy `where` function:

```python
dr_co2_GB = dr_total.where(mask == 143)
dr_co2_GB
```

Check everything went well by repeating the first plot with the selected region:

```python
# choose a good projection for regional maps
proj=ccrs.PlateCarree()

ax = plt.subplot(111, projection=proj)

dr_co2_GB.plot.pcolormesh(ax=ax, transform=ccrs.PlateCarree(), cmap=WhGrYlRd, cbar_kwargs={'shrink': 0.8, 'label': 'Gg/yr'})

ax.coastlines()

# fine tune the extent
ax.set_extent([-11, 36, 36, 64], crs=ccrs.PlateCarree());
```

Looks good - let's take the area average and plot the time series. (Note: you should use `cos(lat)` weights to correctly calculate an area average. Unfortunately this is not yet (as of version 0.10) implemented in xarray.) 

```python
co2_GB_total = dr_co2_GB.sum(dim=('lat', 'lon'))

print(co2_GB_total) # correct

#co2_seasonal_cycle = dr.mean(dim=('lat', 'lon'))*40 # scale it up a bit
# and the line plot
#co2_GB_seasonal_cycle.plot.line(label='GB')
#co2_seasonal_cycle.plot(label='Global')

#plt.legend(ncol=2);
```

## Select using `groupby`

```python
# xarray version > 0.8 is required
xr.__version__
```

```python
# you can group over all integer values of the mask
# you have to take the sum over `stacked_lat_lon`
co2_all = dr_total.groupby(mask).sum('stacked_lat_lon')
co2_all
```

we can add the abbreviations and names of the regions to the DataArray

```python
# extract the abbreviations and the names of the regions from regionmask
abbrevs = regionmask.defined_regions.natural_earth.countries_110[co2_all.region.values].abbrevs
names = regionmask.defined_regions.natural_earth.countries_110[co2_all.region.values].names

co2_all.coords['abbrevs'] = ('region', abbrevs)
co2_all.coords['names'] = ('region', names)
co2_all
#regionmask.defined_regions.natural_earth.countries_110.names
```

now we can select the regions in many ways

```python
#f, axes = plt.subplots(5, 1, figsize=[8, 15])

# as before, by the index of the region
#ax = axes[0]
print(co2_all.isel(region=(co2_all.names == 'United Kingdom')).sum())

#ax = axes[1]
print(co2_all.isel(region=(co2_all.names == 'France')).sum())

# with the long name
#ax = axes[2]
print(co2_all.isel(region=(co2_all.names == 'Spain')).sum())

# with the long name
#ax = axes[3]
print(co2_all.isel(region=(co2_all.names == 'Germany')).sum())

# with the long name
#ax = axes[4]
print(co2_all.isel(region=(co2_all.names == 'Poland')).sum())

#f.tight_layout()
```

```python
fig = plt.figure(figsize=[30, 6])
plt.bar(names,co2_all.values)
plt.xticks(rotation=90)

fig.tight_layout()
fig.savefig('ODIAC_ffco2_countries_bars.png', dpi=300)
```

## selected countries 

```python
countries = ['Germany', 'United Kingdom', 'Italy', 'France', 'Poland', 'Spain']
co2_TNO = [791558, 414196, 353471, 343866, 310313, 269328]
co2_emi = np.arange(len(countries))
for count, name in enumerate(countries):
    co2_emi[count] = co2_all.isel(region=(co2_all.names == countries[count]))
co2_emi
```
